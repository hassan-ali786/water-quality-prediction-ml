```python
# Water Quality EDA Notebook

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
import os
import pandas as pd

base_path = os.path.dirname(os.path.dirname(__file__))
data_path = os.path.join(base_path, "data", "water_potability.csv")

df = pd.read_csv(data_path)

# Basic Info
print("Shape:", df.shape)
print(df.info())
print(df.describe())

# Missing Values
print("\nMissing Values:\n", df.isnull().sum())

# Fill missing values
df.fillna(df.mean(), inplace=True)

# Target distribution
plt.figure()
sns.countplot(x='Potability', data=df)
plt.title('Water Potability Distribution')
plt.show()

# Insight
# 0 = Not Safe, 1 = Safe

# Correlation Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()

# Histogram plots
df.hist(figsize=(12,10))
plt.suptitle('Feature Distributions')
plt.show()

# Boxplots for outliers
plt.figure(figsize=(12,8))
for i, col in enumerate(df.columns[:-1]):
    plt.subplot(3,3,i+1)
    sns.boxplot(y=df[col])
    plt.title(col)
plt.tight_layout()
plt.show()

# Potability vs Features
features = df.columns[:-1]

plt.figure(figsize=(15,10))
for i, col in enumerate(features):
    plt.subplot(3,3,i+1)
    sns.boxplot(x='Potability', y=col, data=df)
plt.tight_layout()
plt.show()

# Pairplot
sns.pairplot(df, hue='Potability')
plt.show()

# Insights
print("""
Key Insights:
1. Dataset had missing values which were filled using mean.
2. Data is slightly imbalanced (more unsafe water samples).
3. Some features like Solids, Conductivity show wide variation.
4. Correlation between features is generally low.
5. Outliers exist in multiple columns.
6. Certain features show different distributions for safe vs unsafe water.
""")
```
